```python
x=[1,2,3,4,5,6]
y=[45,35,67,92,21,41]
z=[9,0,7,22,56,87]
```

```python
pit.plot(x,y)
pit.plot(x,z)
pit.xlabel("Colin")
pit.ylabel("Richard")
pit.show()
```

```python
pit.plot(x,y,"o:r",ms=20,mec='b')
pit.xlabel("Richard")
pit.ylabel("Adonis")
pit.show()
```

```python
pit.plot(x,y,"o:r",ms=20,mec='b')
pit.xlabel("Richard",fontdict = {'color':'blue','size':20})
pit.ylabel("Adonis")
pit.show()
```

```python
pit.plot(x,y)
pit.plot(x,z)
pit.xlabel("Colin",fontdict = {'family':'serif','color':'blue','size':20})
pit.ylabel("Richard",fontdict = {'family':'serif','color':'green','size':30})
pit.show()
```

```python
pit.plot(x,y)
pit.plot(x,z)
pit.xlabel("Colin",fontdict = {'family':'serif','color':'blue','size':20})
pit.ylabel("Richard",fontdict = {'family':'serif','color':'green','size':30})
pit.title("Relationship between Richard and Colin")
pit.show()
```
